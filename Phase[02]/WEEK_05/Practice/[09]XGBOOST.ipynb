# XGBoost

Different machine learning algorithms perform differently depending on the data.

Examples:

- Linear Regression works well for linear data
- Naive Bayes works well for text data

Some algorithms such as:

- Random Forest
- SVM

work reasonably well on a wide variety of datasets, but they may:

- become slow on very large datasets
- consume large memory
- overfit in some situations

---

# XGBoost

XGBoost stands for:

## eXtreme Gradient Boosting

It is one of the most popular machine learning libraries for structured/tabular data.

---

## Important

XGBoost is **not a new machine learning algorithm**.

It is a library that implements:

## Gradient Boosting

with several optimizations.

Thus:

$$
\text{XGBoost}
=
\text{Optimized Gradient Boosting}
$$

---

# Why Gradient Boosting Became Popular

Before understanding XGBoost, it is important to know why Gradient Boosting was widely used.

---

## 1. Flexible

Gradient Boosting can work with many loss functions.

It can be used for:

- Classification
- Regression
- Ranking Problems

---

## 2. Good Performance

Gradient Boosting often provides very strong predictive performance and wins many machine learning competitions.

---

## 3. Robust

It can handle:

- noisy data
- non-linear relationships
- missing values (with suitable implementations)

---

## 4. Popular Model

Because of its flexibility and performance, Gradient Boosting became one of the most widely used ensemble techniques.

---

# Motivation Behind XGBoost

Although Gradient Boosting was powerful, it had some limitations:

- training was slow
- large datasets required significant computation
- overfitting could occur
- memory usage could be high

XGBoost was developed to solve these problems while keeping the strengths of Gradient Boosting.

# XGBoost Features

XGBoost is widely used because it improves both performance and speed compared to traditional Gradient Boosting.

It has three major strengths:

---

# 1. Performance

## Regularized Learning Objective

XGBoost adds regularization to the objective function:

- helps reduce overfitting
- improves generalization

---

## Handling Missing Values

XGBoost can automatically handle missing values.

It learns:

- the best direction (left or right split) for missing values

---

## Sparsity-Aware Split Finding

Efficiently handles sparse data (many zero or missing values).

- skips unnecessary computations
- focuses only on non-zero entries

---

## Efficient Split Finding

Instead of checking all possible splits blindly:

- uses optimized algorithms to find best splits faster
- reduces computation cost significantly

---

## Tree Pruning

XGBoost uses advanced pruning techniques:

- removes unnecessary branches
- prevents over-complex trees
- improves generalization

---

# 2. Speed

## Parallel Processing

Although boosting is sequential across trees:

- split finding inside a tree is parallelized
- improves training speed significantly

---

## Optimized Data Structures

Unlike row-based storage used in many models:

- XGBoost uses column block structure

This means:

- features are stored column-wise
- enables faster split computation and parallel processing

---

## Cache Awareness

XGBoost uses histogram-based training:

- continuous values are bucketed into histograms
- histograms are stored efficiently in CPU cache

This improves memory access speed.

---

## Out-of-Core Computing

Used when dataset is larger than RAM:

- data is split into chunks
- chunks are loaded sequentially into memory

This allows training on very large datasets.

---

## Distributed Computing

Training can be distributed across multiple machines:

- data and computation are split
- each node trains partial model
- master node aggregates results

Tools like:

- Kubernetes
- Dask

can be used for scaling.

---

## GPU Support

XGBoost supports GPU acceleration:

- faster split computation
- efficient parallel execution on GPU cores

---

# 3. Flexibility

## Cross Platform

Works on multiple operating systems:

- Windows
- Linux
- macOS

---

## Multi-Language Support

XGBoost supports many programming languages:

- Python
- R
- Java
- C++
- C
- Ruby

---

## Integration

Easily integrates with other ML tools and frameworks:

- Scikit-learn
- Pandas

- Dask

---

## Supports Many ML Tasks

XGBoost can be used for:

- Regression

# LightGBM

LightGBM (Light Gradient Boosting Machine) is another highly optimized implementation of Gradient Boosting.

It is designed to be:

- faster than traditional Gradient Boosting
- memory efficient
- scalable to very large datasets

---

## Key Idea

LightGBM improves training speed by using:

### Histogram-based learning

Instead of checking every possible split value:

- continuous values are bucketed into histograms
- splits are chosen from histogram bins

This makes training much faster and less memory-heavy.

---

## Special Feature: Leaf-wise Tree Growth

Unlike traditional level-wise tree growth, LightGBM uses:

- leaf-wise (best-first) growth

This means:

- it splits the leaf with maximum loss reduction
- leads to faster convergence

However:

- it can overfit on small datasets if not controlled

---

## Advantages of LightGBM

- very fast training speed
- low memory usage
- good accuracy on large datasets
- handles large feature spaces efficiently

---

# CatBoost

CatBoost is another Gradient Boosting library developed by Yandex.

It is especially powerful for:

- categorical data
- structured/tabular data

---

## Key Idea

CatBoost handles categorical features automatically using:

- ordered boosting
- target encoding techniques

This avoids manual encoding like:

- one-hot encoding
- label encoding

---

## Advantages of CatBoost

- excellent handling of categorical variables
- reduces overfitting using ordered boosting
- strong default performance (less tuning required)
- stable results across datasets

---

# Summary

| Library   | Key Strength |
|----------|-------------|
| XGBoost  | General-purpose, highly optimized boosting |
| LightGBM  | Extremely fast, memory efficient |
| CatBoost  | Best for categorical-heavy datasets |

---

# Final Insight

All three are implementations of Gradient Boosting, but they optimize different aspects:

- XGBoost → performance + scalability
- LightGBM → speed + efficiency
- CatBoost → categorical data + robustness

# XGBoost Regression

Let’s understand XGBoost regression using a simple problem:

Predict salary from GPA

This is similar to Gradient Boosting, but with optimized tree construction and regularization.

---

# Step 1: Initial Prediction

We start by predicting a constant value:

## Mean of target values

$$
\hat{y}_0 = \text{mean}(y)
$$

This becomes the initial baseline prediction.

---

# Step 2: Compute Residuals

We calculate error between actual and predicted values:

$$
\text{Residual} = y - \hat{y}_0
$$

These residuals represent what the model has not learned yet.

---

# Step 3: Train First Tree

A Decision Tree is trained where:

- Input = features (GPA, etc.)
- Target = residuals

This tree learns to predict errors of the baseline model.

---

## First Update Rule

Final prediction becomes:

$$
\hat{y} = \hat{y}_0 + \eta \cdot T_1(x)
$$

where:

- \(T_1(x)\) = output of first tree
- \(\eta\) = learning rate

---

# Step 4: Add More Trees

Now we compute new residuals:

$$
\text{Residual}_2 = y - \hat{y}
$$

Then train another tree on these residuals.

Final prediction becomes:

$$
\hat{y}
=
\hat{y}_0
+
\eta T_1(x)
+
\eta T_2(x)
$$

This continues for multiple trees.

---

# Key Difference in XGBoost Trees

XGBoost does NOT use standard decision tree splitting (like Gini/Entropy).

Instead, it uses a **custom regression tree objective**.

---

# XGBoost Tree Construction

Each leaf contains residuals.

We compute a **Similarity Score**:

$$
SS = \frac{\left(\sum \text{residuals}\right)^2}{n + \lambda}
$$

where:

- \(n\) = number of residuals in the leaf
- \(\lambda\) = regularization term

---

## Intuition of Similarity Score

- higher score → better leaf quality
- encourages compact and meaningful splits
- reduces overfitting using regularization

---

# Split Selection

We try different splits and compute:

## Gain

$$
\text{Gain} = SS_{left} + SS_{right} - SS_{root}
$$

---

## Decision Rule

Choose the split with:

- maximum gain

This ensures best improvement in model performance at every split.

---

# Final Insight

XGBoost regression improves Gradient Boosting by:

- using residual-based learning
- applying regularized leaf scoring
- optimizing split selection using gain
- building trees that reduce loss efficiently

---

# Summary

- Start with mean prediction
- Compute residuals
- Train trees on residuals
- Combine predictions additively
- Use similarity score + gain for better splits

# XGBoost Classification

Suppose we want to predict whether a person is employed or not based on GPA.

Target:

- 1 = Employed
- 0 = Not Employed

Unlike XGBoost Regression, XGBoost Classification works in terms of:

## Log Odds

instead of directly predicting probabilities.

---

# Step 1: Initial Prediction

First compute the probability of positive class:

$$
p = \frac{\text{No. of Positive Samples}}{\text{Total Samples}}
$$

Then calculate:

$$
\text{log odds}
=
\log\left(\frac{p}{1-p}\right)
$$

This becomes the initial prediction.

---

## Converting Log Odds to Probability

Whenever we need probability:

$$
p
=
\frac{e^{\text{log odds}}}
{1 + e^{\text{log odds}}}
$$

The probability is appended to the table.

---

# Step 2: Calculate Pseudo Residuals

Pseudo residuals are:

$$
\text{Residual}
=
\text{Actual Label}
-
\text{Predicted Probability}
$$

or

$$
r_i
=
y_i - p_i
$$

These residuals become the target for the next tree.

---

# Step 3: Train First Decision Tree

Train a Decision Tree using:

- Input = Features
- Target = Residuals

Unlike normal decision trees, XGBoost uses a custom splitting criterion.

---

## Similarity Score

For each node:

$$
SS
=
\frac{
\left(
\sum \text{residual}_i
\right)^2
}
{
\sum p_i(1-p_i)
+
\lambda
}
$$

where:

- \(p_i\) = previous predicted probability
- \(\lambda\) = regularization parameter

A larger Similarity Score indicates a better leaf.

---

## Gain

To evaluate a split:

$$
\text{Gain}
=
(SS_L + SS_R)
-
SS_{\text{root}}
$$

where:

- \(SS_L\) = similarity score of left child
- \(SS_R\) = similarity score of right child

The split with highest gain is selected.

---

# Step 4: Calculate Leaf Output

After constructing the tree, each leaf gets an output value:

$$
\text{Output}
=
\frac{
\sum \text{residual}_i
}
{
\sum p_i(1-p_i)
+
\lambda
}
$$

This value represents the correction suggested by that leaf.

---

# Step 5: Update Prediction

The prediction is updated in log-odds space:

$$
\text{New Log Odds}
=
\text{Old Log Odds}
+
\eta \times
\text{Tree Output}
$$

where:

$$
\eta
$$

is the learning rate.

---

## Convert Back to Probability

The updated log odds are converted to probability:

$$
p
=
\frac{e^{\text{log odds}}}
{1 + e^{\text{log odds}}}
$$

This updated probability is appended to the table.

---

# Step 6: Add Another Tree

Now compute new residuals:

$$
r_i
=
y_i - p_i
$$

using the updated probabilities.

Train another Decision Tree:

- Input = Features
- Target = New Residuals

Again calculate:

- Similarity Scores
- Gain
- Leaf Outputs

---

# Prediction with Two Trees

$$
\text{Log Odds}
=
\text{Initial Log Odds}
+
\eta \cdot T_1(x)
+
\eta \cdot T_2(x)
$$

where:

- \(T_1(x)\) = output of Tree 1
- \(T_2(x)\) = output of Tree 2

---

# Prediction with Three Trees

Extending the same idea:

$$
\text{Log Odds}
=
\text{Initial Log Odds}
+
\eta \cdot T_1(x)
+
\eta \cdot T_2(x)
+
\eta \cdot T_3(x)
$$

---

# Final Probability

After all trees are added:

$$
p
=
\frac{
e^{\text{Final Log Odds}}
}
{
1 + e^{\text{Final Log Odds}}
}
$$

---

# Final Classification

$$
\hat{y}
=
\begin{cases}
1, & p > 0.5 \\
0, & p \le 0.5
\end{cases}
$$

---

# Key Difference from XGBoost Regression

| XGBoost Regression | XGBoost Classification |
|-------------------|------------------------|
| Predicts actual values | Predicts log odds |
| Residual = Actual − Prediction | Residual = Actual − Probability |
| Output directly added to prediction | Output added to log odds |
| Final output is numeric value | Final output is probability/class |

Thus XGBoost Classification performs boosting in **log-odds space**, repeatedly correcting classification errors using residuals and converting the final log odds into probabilities.
