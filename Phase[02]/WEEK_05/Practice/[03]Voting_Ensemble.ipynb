# Voting Ensemble

In Voting Ensemble:

- same dataset is passed to multiple models
- models may be same or different algorithms

Example:

![](./images2/img7.png)

---

# Important Point

The base models are independent of each other.

Each model:

- trains separately
- makes its own prediction

---

# Final Prediction

## Classification

Final output is decided using majority voting.

Example:

- 3 models predict YES
- 2 models predict NO

Final answer:

$$
YES
$$

---

## Regression

For regression tasks, final output is mean of predictions:

$$
\hat{y}
=
\frac{1}{n}
\sum_{i=1}^{n} \hat{y}_i
$$

---

# Types of Voting

## Hard Voting

Uses final class labels for voting.

Example:

- YES
- NO
- YES

Final prediction:

$$
YES
$$

---

## Soft Voting

Uses prediction probabilities.

The probabilities are averaged and class with highest probability is selected.

Soft voting generally performs better.

## Classification Code

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
df = pd.read_csv("./dataset/Iris.csv")
```

```python
df.head()
```

```python
df = df.iloc[:,1:]
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
df["Species"] = encoder.fit_transform(df["Species"])
```

```python
df.head()
```

```python
new_df = df[df['Species'] != 0][['SepalLengthCm','SepalWidthCm','Species']]
```

```python
new_df.head()
```

```python
X = df.drop(columns="Species")
y = df["Species"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
model_1 = LogisticRegression()
model_2 = RandomForestClassifier()
model_3 = KNeighborsClassifier()
```

```python
estimators = [('lr',model_1),('rf', model_2),('knn',model_3)]
```

```python
for estimator in estimators:
    x = cross_val_score(estimator[1],X,y,cv=10,scoring='accuracy')
    print(estimator[0],np.round(np.mean(x),2))
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc = VotingClassifier(estimators=estimators,voting='hard')
x = cross_val_score(vc,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

```python
vc1 = VotingClassifier(estimators=estimators,voting='soft')
x = cross_val_score(vc1,X,y,cv=10,scoring='accuracy')
print(np.round(np.mean(x),2))
```

## Voting Regression

```python
from sklearn.datasets import fetch_california_housing
import numpy as np
```

```python
data = fetch_california_housing()
X = data.data
y = data.target
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
svr = SVR()
```

```python
estimators = [('lr',lr),('dt',dt),('svr',svr)]
```

```python
for estimator in estimators:
  scores = cross_val_score(estimator[1],X,y,scoring='r2',cv=10)
  print(estimator[0],np.round(np.mean(scores),2))    
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```

```python
dt1 = DecisionTreeRegressor(max_depth=1)
dt2 = DecisionTreeRegressor(max_depth=3)
dt3 = DecisionTreeRegressor(max_depth=5)
dt4 = DecisionTreeRegressor(max_depth=7)
dt5 = DecisionTreeRegressor(max_depth=None)
```

```python
estimators = [('dt1',dt1),('dt2',dt2),('dt3',dt3),('dt4',dt4),('dt5',dt5)]
```

```python

for estimator in estimators:
  scores = cross_val_score(estimator[1],X,y,scoring='r2',cv=10)
  print(estimator[0],np.round(np.mean(scores),2))
```

```python
vr = VotingRegressor(estimators)
scores = cross_val_score(vr,X,y,scoring='r2',cv=10)
print("Voting Regressor",np.round(np.mean(scores),2))
```
