# Ensemble Techniques

Ensemble learning is the collection of multiple models (base models) to solve a problem.

It is based on the idea of:

## Wisdom of the Crowd

Multiple weak/moderate models together often perform better than a single model.

---

# Prediction

Suppose we have models:

$$
m_1, m_2, m_3, \dots, m_n
$$

These models may:

- be different algorithms  
  (Linear Regression, KNN, Decision Tree, etc.)

or

- be same algorithm trained on different data

---

## Classification

Input is sent to all models.

Final output is decided using majority voting.

Example:

- 3 out of 5 models predict YES
- Final prediction = YES

---

## Regression

Input is sent to all models.

Final prediction is calculated using mean:

$$
\hat{y}
=
\frac{1}{n}
\sum_{i=1}^{n} \hat{y}_i
$$

---

# Training

Training process depends on type of ensemble method.

# Types of Ensemble Techniques

## 1) Voting

Base models are different algorithms.

Prediction is based on:

- voting for classification
- mean for regression

## 2) Bagging

Bagging stands for:

## Bootstrap Aggregation

- Base models are same
- Datasets are different

Different datasets are created using random sampling with replacement.

Example:

![](./images2/img5.png)

Example algorithm:

- Random Forest

(Random Forest uses Decision Trees as base models)

---

## 3) Boosting

Boosting is one of the most powerful ensemble methods.

- Base models are same
- Models are trained sequentially

Idea:

- First model makes predictions
- Next model focuses more on previous mistakes
- Errors are gradually reduced

Example:

![](./images2/img6.png)

Algorithms:

- AdaBoost
- Gradient Boosting
- XGBoost

---

## 4) Stacking

In stacking:

- outputs of base models are fed into another model

This final model is called:

- meta model
- blender model

The meta model learns:

- which models are better
- how much importance each model should get
