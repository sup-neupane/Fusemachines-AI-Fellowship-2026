## Polynomial Logistic Regression

This is used when the data cannot be separated using a straight line (linear boundary).

![](./images/img9.png)

---

## Idea

We extend logistic regression by adding polynomial features, similar to polynomial linear regression.

Instead of only using original features, we create higher-order terms.

---

## Feature Transformation

For example, if we set degree = 2, we expand features as:

- x₁, x₂, ..., xₙ  
- x₁², x₂², ..., xₙ²  
- interaction terms like x₁x₂, x₂x₃, etc.

So the dataset becomes richer and non-linear patterns can be captured.

---

## Model Form

After transformation, logistic regression is applied:

$$
z = w_0 + w_1 x_1 + w_2 x_2 + w_3 x_1^2 + w_4 x_2^2 + w_5 x_1 x_2 + \dots
$$

$$
\hat{y} = \sigma(z)
$$

---

## Intuition

- Original logistic regression → linear decision boundary  
- Polynomial logistic regression → curved / non-linear decision boundary  

This allows the model to separate complex datasets that are not linearly separable.

---

## Key Insight

We are not changing logistic regression itself —  
we are just **increasing feature space complexity**, then applying standard logistic regression on top.
