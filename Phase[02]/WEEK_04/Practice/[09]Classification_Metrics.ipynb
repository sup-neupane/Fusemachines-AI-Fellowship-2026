# Classification Metrics

## Accuracy (Binary Classification)

Accuracy measures how many predictions were correct out of total predictions.

$$
\text{Accuracy} =
\frac{\text{No of Correct Predictions}}
{\text{Total Predictions}}
$$

or

$$
\text{Accuracy} =
\frac{TP + TN}
{TP + TN + FP + FN}
$$

---

## Accuracy (Multi Class Classification)

For multiclass classification also:

$$
\text{Accuracy} =
\frac{\text{No of Correct Predictions}}
{\text{Total Predictions}}
$$

---

The desired accuracy depends on the dataset and problem statement.

For some applications even 70% may be good, while for critical systems like medical diagnosis very high accuracy is required.

---

## Limitation of Accuracy

Accuracy alone is not always a good metric because it does not tell us the type of error made by the model.

For imbalanced datasets, accuracy can be misleading.

# Confusion Matrix

![](./images/img8.png)

A confusion matrix shows:

- TP → True Positive  
  Model predicted positive and actual class was also positive

- TN → True Negative  
  Model predicted negative and actual class was also negative

- FP → False Positive  
  Model predicted positive but actual class was negative

- FN → False Negative  
  Model predicted negative but actual class was positive

---

## Accuracy Using Confusion Matrix

$$
\text{Accuracy} =
\frac{TP + TN}
{TP + TN + FP + FN}
$$

---

## Type of Errors

### Type 1 Error

False Positive (FP)

Model predicts positive when actual class is negative.

---

### Type 2 Error

False Negative (FN)

Model predicts negative when actual class is positive.

---

## Multiclass Confusion Matrix

Confusion matrices can also be created for multiclass classification problems.

In multiclass confusion matrices:

- diagonal elements represent correct predictions
- non-diagonal elements represent wrong classifications

## Note: Accuracy is misleading for imbalanced dataset

### Example

Suppose:

- 100 samples total  
- 95 are class 0  
- 5 are class 1  

A model predicts **all samples as class 0**.

---

### Result:

- Correct = 95  
- Incorrect = 5  

$$
\text{Accuracy} = \frac{95}{100} = 95\%
$$

---

### Problem:

Model completely fails to detect class 1, even though accuracy looks very high.

## Precision

Precision tells what percentage of predicted positives are actually correct.

$$
\text{Precision} = \frac{TP}{TP + FP}
$$

It is important when **False Positives (FP)** are costly.

### Example:
Spam detection → marking a real email as spam is bad.

---

## Recall

Recall tells what percentage of actual positives are correctly identified.

$$
\text{Recall} = \frac{TP}{TP + FN}
$$

It is important when **False Negatives (FN)** are costly.

### Example:
Cancer detection → missing a cancer case is very dangerous.

---

## F1 Score

F1 score balances Precision and Recall.

$$
F1 = \frac{2PR}{P + R}
$$

It is the harmonic mean of Precision and Recall.

### Insight:
Harmonic mean penalizes low values more, so:
- if either Precision or Recall is low → F1 score becomes low

---

## Multi-Class Classification

For multi-class problems, Precision, Recall, and F1 are computed using two main approaches:

### 1. Macro Average
- Compute metric for each class separately
- Take simple average

$$
\text{Macro Precision} = \frac{P_1 + P_2 + \dots + P_k}{k}
$$

(similarly for Recall and F1)

---

### 2. Weighted Average
- Each class metric is weighted by number of samples

$$
\text{Weighted Precision} = \sum_{i=1}^{k} w_i P_i
$$

where \(w_i\) = proportion of class i samples

---

### 3. Micro Average
- Compute global TP, FP, FN first
- Then calculate metrics

This treats all classes equally at instance level.
