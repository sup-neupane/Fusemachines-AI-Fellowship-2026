# Elastic Net Regression

Elastic Net Regression is a combination of **Ridge Regression (L2)** and **Lasso Regression (L1)**.

---

## Ridge (L2)

$$
L = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda w^2
$$

---

## Lasso (L1)

$$
L = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda |w|
$$

---

## When to Use

- Ridge is used when we **do not want to remove any features**
- Lasso is used for **feature selection** (some features may be removed)

Elastic Net is used when we are **not sure whether to use Ridge or Lasso**.

---

## Elastic Net Loss Function

$$
L = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \alpha w^2 + \beta |w|
$$

---

## Hyperparameters

- \(\lambda = \alpha + \beta\)
- L1 ratio:
  $$
  \frac{\alpha}{\alpha + \beta}
  $$

---

## Key Idea

By adjusting \(\alpha\) and \(\beta\), we can control the balance between:

- Ridge behavior (L2 regularization)
- Lasso behavior (L1 regularization)

This makes Elastic Net flexible and more robust in real-world datasets.

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X,y = load_diabetes(return_X_y=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Linear Regression
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Ridge 
reg = Ridge(alpha=0.1)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# Lasso
reg = Lasso(alpha=0.01)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```

```python
# ElasticNet
reg = ElasticNet(alpha=0.005,l1_ratio=0.9)
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
r2_score(y_test,y_pred)
```
