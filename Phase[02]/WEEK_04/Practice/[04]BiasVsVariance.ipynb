# Bias-Variance Tradeoff

---

## Bias

Bias is the **inability of a machine learning model to capture the true relationship in the data**.

- High bias → model is too simple
- Leads to underfitting

---

## Variance

Variance is the **sensitivity of the model to changes in the training data**.

- High variance → model changes a lot with different datasets
- Leads to overfitting

---

## Overfitting

- Model performs very well on training data
- Fails to generalize on unseen data

---

## Underfitting

- Model fails to capture patterns in training data
- Performs poorly on both training and test data

---

## Goal

We aim for:

> Low bias + low variance
