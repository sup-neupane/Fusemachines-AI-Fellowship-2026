# Naive Bayes

## Conditional Probability

If there are two events \(A\) and \(B\), then probability of \(A\) given \(B\) is:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

given that:

$$
P(B) \ne 0
$$

---

## Dice Example

Suppose we roll two dice:

- \(D_1\)
- \(D_2\)

Total sample space:

$$
6 \times 6 = 36
$$

Sample space:

$$
\{
(1,1), (1,2), \dots , (6,6)
\}
$$

---

## Sum Table

| D1 \\ D2 | 1 | 2 | 3 | 4 | 5 | 6 |
|---|---|---|---|---|---|---|
| 1 | 2 | 3 | 4 | 5 | 6 | 7 |
| 2 | 3 | 4 | 5 | 6 | 7 | 8 |
| 3 | 4 | 5 | 6 | 7 | 8 | 9 |
| 4 | 5 | 6 | 7 | 8 | 9 | 10 |
| 5 | 6 | 7 | 8 | 9 | 10 | 11 |
| 6 | 7 | 8 | 9 | 10 | 11 | 12 |

---

### Example 1

What is probability that \(D_1 = 5\)?

Possible outcomes:

$$
(5,1),(5,2),(5,3),(5,4),(5,5),(5,6)
$$

Total favorable outcomes = 6

So:

$$
P(A) = \frac{6}{36} = \frac{1}{6}
$$

---

### Example 2

What is probability that:

$$
D_1 + D_2 \le 10
$$

Only sums 11 and 12 are excluded.

- Sum = 11 → 2 outcomes  
- Sum = 12 → 1 outcome  

Excluded outcomes = 3

So favorable outcomes:

$$
36 - 3 = 33
$$

Therefore:

$$
P(B) = \frac{33}{36} = \frac{11}{12}
$$

---

### Example 3

What is probability:

$$
P(D_1 = 5 \mid D_1 + D_2 \le 10)
$$

Let:

- \(A\): \(D_1 = 5\)
- \(B\): \(D_1 + D_2 \le 10\)

Using conditional probability:

$$
P(A \mid B) = \frac{P(A \cap B)}{P(B)}
$$

Now:

Valid outcomes for \(A \cap B\):

$$
(5,1),(5,2),(5,3),(5,4),(5,5)
$$

Total = 5 outcomes

So:

$$
P(A \cap B) = \frac{5}{36}
$$

Therefore:

$$
P(A \mid B)
=
\frac{5/36}{33/36}
=
\frac{5}{33}
$$

---

# Independent Events

Two events are independent if:

$$
P(A \cap B) = P(A)P(B)
$$

Meaning:

- occurrence of one event does not affect the other

---

# Mutually Exclusive Events

Two events are mutually exclusive if:

$$
P(A \cap B) = 0
$$

Meaning:

- both events cannot happen together

### Example:
Getting Head and Tail at the same time in a single coin toss.

# Bayes Theorem

If there are two events \(A\) and \(B\), then:

$$
P(A \mid B)
=
\frac{P(B \mid A)P(A)}{P(B)}
$$

given that:

$$
P(B) \ne 0
$$

---

## Terms

- \(P(A \mid B)\) → Posterior  
- \(P(B \mid A)\) → Likelihood  
- \(P(A)\) → Prior  
- \(P(B)\) → Evidence  

---

# Example

Suppose a factory has three machines:

- \(M_1\) produces 20% markers
- \(M_2\) produces 30% markers
- \(M_3\) produces 50% markers

Defect rates:

- \(M_1\) → 5% defective
- \(M_2\) → 3% defective
- \(M_3\) → 1% defective

---

## Problem

If a randomly selected marker is defective, what is the probability that it was produced by \(M_3\)?

We need to calculate:

$$
P(M_3 \mid D)
$$

where:

- \(D\) = defective marker

---

## Step 1: Apply Bayes Theorem

$$
P(M_3 \mid D)
=
\frac{P(D \mid M_3)P(M_3)}{P(D)}
$$

---

## Step 2: Substitute Known Values

From the question:

$$
P(D \mid M_3) = 0.01
$$

$$
P(M_3) = 0.5
$$

---

## Step 3: Calculate Evidence \(P(D)\)

Using total probability:

$$
P(D)
=
P(D \mid M_1)P(M_1)
+
P(D \mid M_2)P(M_2)
+
P(D \mid M_3)P(M_3)
$$

$$
=
(0.05)(0.2)
+
(0.03)(0.3)
+
(0.01)(0.5)
$$

$$
=
0.01 + 0.009 + 0.005
$$

$$
=
0.024
$$

---

## Step 4: Final Calculation

$$
P(M_3 \mid D)
=
\frac{(0.01)(0.5)}{0.024}
$$

$$
=
\frac{0.005}{0.024}
$$

$$
\approx 0.2083
$$

---

## Final Answer

$$
P(M_3 \mid D) \approx 20.83\%
$$

So if a marker is defective, there is about **20.83% probability** that it was produced by machine \(M_3\).

# Naive Bayes Classifier

Suppose we want to predict whether a team will:

- Win (\(W\))
- Lose (\(L\))

based on input features like:

- toss
- venue
- outlook

This is a two-class classification problem.

---

## Goal

We calculate:

$$
P(L \mid \{\text{toss, venue, outlook}\})
$$

and

$$
P(W \mid \{\text{toss, venue, outlook}\})
$$

Using Bayes theorem:

$$
P(W \mid X)
=
\frac{P(X \mid W)P(W)}{P(X)}
$$

$$
P(L \mid X)
=
\frac{P(X \mid L)P(L)}{P(X)}
$$

where:

$$
X = \{\text{toss, venue, outlook}\}
$$

---

## Important Observation

The denominator:

$$
P(X)
$$

is same for all classes.

So while comparing probabilities, we ignore it.

Thus:

$$
P(W \mid X) \propto P(X \mid W)P(W)
$$

$$
P(L \mid X) \propto P(X \mid L)P(L)
$$

---

# Why is it called "Naive"?

Naive Bayes assumes that all input features are independent given the class.

So:

$$
P(\{\text{toss, venue, outlook}\} \mid W)
$$

is simplified as:

$$
P(\text{toss} \mid W)
P(\text{venue} \mid W)
P(\text{outlook} \mid W)
$$

This independence assumption is called the **naive assumption**.

---

## General Formula

$$
P(c_k \mid \mathbf{x})
\propto
P(c_k)
\prod_{i=1}^{n} P(x_i \mid c_k)
$$

where:

- \(P(c_k)\) → prior probability of class
- \(P(x_i \mid c_k)\) → likelihood
- \(x_i\) → input feature

---

## Final Prediction Formula

Naive Bayes calculates probability for every class and selects the class with maximum probability.

$$
\hat{y}
=
\underset{k \in \{1,2,\dots,K\}}{\operatorname{argmax}}
\;
P(c_k)
\prod_{i=1}^{n} P(x_i \mid c_k)
$$

---

## Key Insight

Naive Bayes is:

- simple
- fast
- works well for text classification and spam detection
- based completely on probability

# Naive Bayes for Numerical Data

## Gaussian Naive Bayes

When input columns are numerical, for example:

- height
- weight

and target is:

- gender

we use **Gaussian Naive Bayes**.

---

## Main Assumption

Gaussian Naive Bayes assumes that numerical features follow a **Normal Distribution (Gaussian Distribution)**.

For every feature and every class, we calculate:

- Mean (\(\mu\))
- Standard Deviation (\(\sigma\))

---

## Gaussian Probability Density Function

Probability of a feature value is calculated using:

$$
P(x_i \mid c_k)
=
\frac{1}{\sqrt{2\pi\sigma_k^2}}
\exp
\left(
-\frac{(x_i - \mu_k)^2}{2\sigma_k^2}
\right)
$$

where:

- \(x_i\) → feature value
- \(\mu_k\) → mean of feature for class \(c_k\)
- \(\sigma_k\) → standard deviation for class \(c_k\)

---

## Workflow

### Step 1

Separate data class-wise.

Example:

- Male
- Female

---

### Step 2

For each feature calculate:

- mean
- standard deviation

for every class separately.

---

### Step 3

Use Gaussian formula to calculate:

$$
P(x_i \mid c_k)
$$

for all features.

---

### Step 4

Apply Naive Bayes formula:

$$
P(c_k \mid \mathbf{x})
\propto
P(c_k)
\prod_{i=1}^{n} P(x_i \mid c_k)
$$

---

### Step 5

Choose class with highest probability.

---

# Other Distributions

If data does not follow normal distribution, other distributions can also be used:

- Binomial Distribution
- Poisson Distribution
- Multinomial Distribution

Different Naive Bayes variants are chosen depending on data distribution and feature type.

---

## Examples

- Gaussian Naive Bayes → continuous numerical data
- Multinomial Naive Bayes → text/count data
- Bernoulli Naive Bayes → binary features
