## Data Wrangling Stages

There are several stages in data wrangling. The first phase is:

## 1. Handling Missing Values

We can handle missing values using two common methods:

- **Dropping incomplete records**  
  Removing rows or columns that contain missing values.

- **Filling missing values**  
  Replacing missing values with an appropriate substitute (e.g., mean, median, or a constant value).

Before deciding how to handle missing values, it is important to assess the data. This helps in choosing the most appropriate method for treatment.

```python
missing_values_count = rewards_df.isna().sum()
missing_values_count
```

```python
percentage_missing_values = missing_values_count/ len(rewards_df)  * 100
percentage_missing_values
```

From the assessment, the columns **birthday**, **city**, **state**, and **zip** have a high percentage of missing values.  

Due to the large amount of missing data, these columns may not be reliable and could be considered for removal or handled with caution.

On the other hand, the **lastSeen** column has a relatively low percentage of missing values.  
Therefore, instead of dropping it, we will try to fill in the missing values using appropriate techniques.

```python
rewards_df['Last Seen']
```

```python
#Convert to datetime as it is string
rewards_df['Last Seen'] = pd.to_datetime(rewards_df['Last Seen'])
```

```python
rewards_df['Last Seen']
```

```python
rewards_df['Last Seen'].max()
```

```python
rewards_df['Last Seen'].min()
```

Since we know the minimum and maximum values of the `lastSeen` column, we can use this information to approximate the missing entries.To handle the missing values, we will fill them using the **minimum value** of the `lastSeen` column. This approach assumes that missing entries correspond to the earliest observed activity.

```python
rewards_df['Last Seen'] = rewards_df['Last Seen'].fillna(rewards_df['Last Seen'].min())
```

```python
missing_values_count = rewards_df.isna().sum()
missing_values_count
```

Since the columns **birthday**, **city**, **state**, and **zip** contain a high percentage of missing values, we will drop them from the dataset.

After removing these columns, we will evaluate whether the remaining data is sufficient to perform meaningful analysis.

```python
rewards_df_sample = rewards_df.dropna()
rewards_df_sample
```

```python
missing_values_count = rewards_df_sample.isna().sum()
missing_values_count
```

```python
#Lets drop tags as it isnt useful
rewards_df_sample = rewards_df_sample.drop(columns='Tags')
rewards_df_sample
```

## Stage 2: Handling Messy and Inconsistent Data

Now we move to the second stage of data wrangling: handling messy or inconsistent data.

This stage involves:

- Correcting inconsistent formatting (e.g., date formats, text casing)  
- Standardizing values (e.g., "NY", "New York", "new york" → "New York")  
- Fixing data entry errors or typos  
- Ensuring uniform data types across columns  

Cleaning inconsistent data helps improve data quality and ensures more accurate analysis.

```python
city_unique_values = rewards_df_sample['City'].unique()
city_unique_values = sorted(city_unique_values)
city_unique_values
```

Here, we observe that the same city names are stored in different formats. This inconsistency can lead to incorrect analysis and grouping.

To address this, we need to standardize the city name.
